```python
# %load youtube_selenium1025.py
from selenium.webdriver import Chrome
from selenium.webdriver.support.ui import Select
import requests
from bs4 import BeautifulSoup
import time
import os
import pandas as pd
from selenium.webdriver.common.keys import Keys    #鍵盤
from selenium.webdriver.support import expected_conditions as EC    #等到出現
import json
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait


#data
videodic={ 'title':'',
    'link':'',
    'date':'',
    'description':'',
    'comment':''}



upath = r'./youtube'

if not os.path.exists(upath):
    os.mkdir(upath)

url = ('https://www.youtube.com/results?search_query=%E4%BF%A1%E7%94%A8%E5%8D%A1')
# 開啟網頁
driver = Chrome('./chromedriver')

driver.get(url)
# 滾動到底
# js="var q=document.documentElement.scrollTop=100000"
# driver.execute_script(js)
#driver.execute_script('window.scrollTo(1, 3000);')
driver.find_element_by_tag_name('html').send_keys(Keys.END)

# 滾回頂
js = "var q=document.documentElement.scrollTop=0"
#driver.execute_script(js)
for i in range(20):
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="video-title"]').send_keys(Keys.END)
#driver.find_element_by_tag_name('html').send_keys(Keys.HOME)
time.sleep(3)
# 若要对页面中的内嵌窗口中的滚动条进行操作，要先定位到该内嵌窗口，在进行滚动条操作
# js="var q=document.getElementById('id').scrollTop=100000"
# driver.execute_script(js)
time.sleep(3)
htmlsource = driver.page_source
driver.quit()
allvideo=[]
soup = BeautifulSoup(htmlsource, 'html.parser')
info = soup.select('div.text-wrapper')
df=pd.DataFrame(columns=['title','link','date','views','description','comment'])
idx=0

for i in info:
    videodata={'title':'',
    'link':'',
    'date':'',
    'views':'',
    'description':'',
    'comment':''}

    title=''
    link=''
    date=''
    views=''
    description=''
    comment=''

    driver = Chrome('./chromedriver')
    title = (i.select('a#video-title')[0].text).strip()  # 標題
    link = 'https://www.youtube.com/' + i.a['href']  # link
    driver.get(link)
    # driver.execute_script(js)
    # m=driver.find_element_by_id("content-text")
    driver.execute_script('window.scrollTo(1, 500);')
    # now wait let load the comments
    time.sleep(2)
    try:
        clickelement=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="more"]')))
        clickelement.click()


        #driver.find_element_by_xpath('//*[@id="more"]').click()
        date = driver.find_element_by_xpath('//*[@id="date"]/yt-formatted-string').text
        views = driver.find_element_by_xpath('//*[@id="count"]/yt-view-count-renderer/span[1]')
        description = driver.find_element_by_xpath('//*[@id="description"]/yt-formatted-string').text
        views=views.text.split('：')[1][:-1]
        #driver.execute_script('window.scrollTo(1, 3000);')
        driver.find_element_by_tag_name('html').send_keys(Keys.END)
        time.sleep(2)
        #comment_div=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="content-text"]')))
        #print(comment_div.text)
        comment_div = driver.find_element_by_xpath('//*[@id="contents"]')
        comments = comment_div.find_elements_by_xpath('//*[@id="content-text"]')
        #print(comments)

        print(title)
        savefile = title + '\n'
        print(views)
        print(date)
        print(link)
        #print(content.text)
        savefile += description +'\n'
        for c in comments:
            #print(comment.text)
            savefile += c.text + '\n'
            comment += c.text+'\n'
            print(c.text)
        #json
        videodata['title']=title
        videodata['link']=link
        videodata['date']=date
        videodata['views']=views
        videodata['description']=description
        videodata['comment']=comment
        allvideo.append(videodata)

        #df
        videodf=[title,link,date,views,description,comment]
        df.loc[idx]=videodf
        idx+=1

        driver.quit()
        print(savefile)
        stopsign =  "<, >, :, \", /, \, |, ?, *"
        for i in title:
            if i in stopsign:
                title = title.replace(i, '-')
        with open(upath+'/{}.txt'.format(title), 'w', encoding='utf-8') as w:
            w.write(savefile)
    except:
        pass
    
```
